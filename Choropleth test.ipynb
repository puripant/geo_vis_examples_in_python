```python
import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world[world.continent != 'Antarctica']
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
```

```python
world.plot(column='continent', figsize=(30, 10))
plt.axis('off')
```

```python
world.plot(column='pop_est', cmap='PuOr', scheme='quantiles', figsize=(30, 10))
plt.axis('off')

world['pop_est']
scale = plt.Normalize().autoscale(world['pop_est'])
legend = plt.cm.ScalarMappable(cmap='PuOr', norm=scale)
legend._A = []
plt.colorbar(legend)
```

```python
base = world.plot(color='lightgray', edgecolor='0.8', linewidth=1, figsize=(30, 10))
cities.plot(ax=base, marker='o', color='red', markersize=10)
plt.axis('off')
```
